  4.  Tasks to Be Performed:

1. Put master.txt on master branch, stage and commit

2. Create 3 branches: public 1, public 2 and private

3. Put public1.txt on public 1 branch, stage and commit

4. Merge public 1 on master branch

5. Merge public 2 on master branch

6. Edit master.txt on private branch, stage and commit

7. Now update branch public 1 and public 2 with new master code in private

8. Also update new master code on master

9. Finally update all the code on the private branch

Solution -->

````python

# 1. Create repo and master.txt in master branch
mkdir MyRepo && cd MyRepo
git init
touch master.txt
git add master.txt
git commit -m

# 2. Create branches public1, public2, private
git branch public1
git branch public2
git branch private

# 3. Add public1.txt in public1 branch
git checkout public1
touch public1.txt
git add public1.txt
git commit -m

# 4. Merge public1 into master
git checkout master
git merge public1

# 5. Add public2.txt in public2 branch and merge into master
git checkout public2
touch public2.txt
git add public2.txt
git commit -m
git checkout master
git merge public2

# 6. Edit master.txt in private branch
git checkout private
echo  >> master.txt
git add master.txt
git commit -m

# 7. Merge private into master
git checkout master
git merge private

# 8. Update public1 with master changes
git checkout public1
git merge master

# 9. Update public2 with master changes
git checkout public2
git merge master

# 10. Merge master into private (final sync)
git checkout private
git merge master


# Push Commands

# After running the previous commands, add GitHub remote
git remote add origin https://github.com/<Tiwnikhil26>

# Push master branch
git push -u origin master

# Push all other branches
git push origin public1
git push origin public2
git push origin private

````

  5. Tasks to Be Performed:

1. Create a Git Flow workflow architecture on Git

2. Create all the required branches

3. starting from e feature branch, push the branch to the master, following the architecture

4. Push an urgent.txt on master using hotfix

Solution -- >
  - Git Flow typically has:

master → production-ready code

develop → integration branch for features

feature/ → branches for new features

release/ → prep branch before going to production

hotfix/ → urgent fixes directly from master

Flow:

    feature → develop → release → master
               ↑                    ↑
               |                    |
             hotfix --------------→ master


````python
# 1. Create repo and init
mkdir GitFlowRepo && cd GitFlowRepo
git init

# 2. Create master branch with initial commit
touch readme.md
echo "Initial Commit" > readme.md
git add readme.md
git commit -m "Initial commit on master"

# 3. Create develop branch from master
git branch develop

# 4. Create a feature branch from develop
git checkout develop
git checkout -b feature/login

# 5. Add a feature file and commit
touch login.txt
git add login.txt
git commit -m "Add login feature"

# 6. Merge feature into develop
git checkout develop
git merge feature/login

# 7. Create a release branch from develop
git checkout -b release/v1.0
touch release_notes.txt
git add release_notes.txt
git commit -m "Prepare release v1.0"

# 8. Merge release into master
git checkout master
git merge release/v1.0

# 9. Merge release back into develop
git checkout develop
git merge release/v1.0

# 10. Create hotfix branch from master for urgent.txt
git checkout master
git checkout -b hotfix/urgent-fix
touch urgent.txt
git add urgent.txt
git commit -m "Add urgent.txt in hotfix"

# 11. Merge hotfix into master
git checkout master
git merge hotfix/urgent-fix

# 12. Merge hotfix back into develop
git checkout develop
git merge hotfix/urgent-fix
````

````python
## Push Everything to GitHub

git remote add origin https://github.com/<your-username>/<your-repo>.git
git push -u origin master
git push origin develop
git push origin feature/login
git push origin release/v1.0
git push origin hotfix/urgent-fix
````
